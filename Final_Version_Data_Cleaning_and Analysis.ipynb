### tokenization, stop word remove, lemmatization

```python
## Since we have 15 text files, I'm creating a function to do tokenization, remove stop words, and lemmatization for text and applying to the dataframe so it automatically helps me process all text at once

def process_text(text):
    # tokenization
    tokens = word_tokenize(text.lower())
    
    # stop words removal
    filtered_tokens = [token for token in tokens if token not in stops]
    
    # Lemmatize tokens
    lemmatized_tokens = [wordnet_lemmatizer.lemmatize(token) for token in filtered_tokens]
    
    # Rejoin tokens into string
    processed_text = ' '.join(lemmatized_tokens)
    return processed_text

# apply the function to the dataframe
ipcc_csv['processed_text'] = ipcc_csv['text'].apply(process_text)

# double-checking
ipcc_csv.head()
```

### To extract the sentence containg the confidence level, the sentence before, and after

```python
def extract(text, phrase):
    sentences = nltk.sent_tokenize(text)

    high_confidence_list = []

    for i, sentence in enumerate(sentences):
        #assigning the phrase
        if phrase in sentence:
            phrase_idx = i
            before_idx = i-1
            after_idx = i+1

            # extract the sentences before, containing, and after the phrase
            before = sentences[before_idx] if before_idx >= 0 else ""
            containing = sentence
            after = sentences[after_idx] if after_idx < len(sentences) else ""

            whole = [before, containing, after]
            high_confidence_list.append(whole)

    return high_confidence_list
```

```python
phrases = ['high confidence', 'medium confidence', 'low confidence']

for i in range(len(phrases)):
    phrase = phrases[i]
    ipcc_csv[f'{phrase}_text'] = ipcc_csv['processed_text'].apply(extract, phrase = phrase)
```

#### to flatten the list of lists into a single list.  This is because I was unable to do tfdif without merging the lists.

```python
def clean(text):
    flat_list = [item for sublist in text for item in sublist]

    # Use map to convert each item in the list to a string
    str_list = map(str, flat_list)

    # Join the list of strings with a space delimiter
    result = ' '.join(str_list)
    return result
```

```python
ipcc_csv['high confidence_text'] = ipcc_csv['high confidence_text'].apply(clean)
ipcc_csv['medium confidence_text'] = ipcc_csv['medium confidence_text'].apply(clean)
ipcc_csv['low confidence_text'] = ipcc_csv['low confidence_text'].apply(clean)
```

### TF - IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
```

#### performing the tf-idf test

```python
def get_tfidf(text):
    
    # Create the TF-IDF vectorizer
    vectorizer = TfidfVectorizer()
    
    feature_list = []
    
    try:
        # Fit and transform the vectorizer on the preprocessed text
        tfidf = vectorizer.fit_transform([text])

        # Get the feature names and the corresponding TF-IDF scores
        feature_names = vectorizer.get_feature_names()
        tfidf_scores = tfidf.toarray()[0]

        # Zip the feature names and the corresponding TF-IDF scores together
        features = list(zip(feature_names, tfidf_scores))

        # Sort the features by their TF-IDF scores since we want to have the score from the highest to the lowest
        features = sorted(features, key=lambda x: x[1], reverse=True)

        # Print the top 30 features with the highest TF-IDF scores
        for feature in features[:30]:
            feature_list.append(feature)

        return feature_list
    
    ## this is because there are some sections without low confidence labels. This is to return an empty list instead of raising an error. 
    except:
        return feature_list
```

```python
ipcc_csv['high_tfidf'] = ipcc_csv['high confidence_text'].apply(get_tfidf)
ipcc_csv['medium_tfidf'] = ipcc_csv['medium confidence_text'].apply(get_tfidf)
ipcc_csv['low_tfidf'] = ipcc_csv['low confidence_text'].apply(get_tfidf)
```

```python
import spacy
from spacy import displacy
nlp = spacy.load("en_core_web_sm")
nlp.max_length = 2500000
```

```python
all_entities = []
for i in range(len(ipcc_csv)):
    text = ipcc_csv.iloc[i,0]
    doc = nlp(text)
    entity_type = [] 
    for ent in doc.ents:
        entity_type.append(ent.label_)
    entity_identified = [] 
    for ent in doc.ents:
        entity_identified.append(ent.text)
    ent_dict = {'file_name': ipcc_csv.iloc[i, 2], 'Entity_type': entity_type, 'Entity_identified': entity_identified}
    all_entities.append(ent_dict)
```
